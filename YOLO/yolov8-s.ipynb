```python
yaml_path = '/kaggle/input/vehicle-detection/data/data.yaml'
```

```python
from ultralytics import YOLO
import os
```

```python
# Load YOLOv8n model
model = YOLO('yolov8s.pt')

# Check model architecture
model.info()
```

```python
model.train(
    data='/kaggle/input/vehicle-detection/data/data.yaml',  # Path to the data.yaml file
    epochs=50,  # Number of epochs
    imgsz=640,  # Image size
    batch=16,  # Batch size
    project='/kaggle/working//vehicle_classification_model',  # Save results in the same directory
    name='vehicle_classification_yolov8n',  # Name of the experiment
    device=0,  # Use GPU
    save_period=1
)
```

```python
results = model.val()
```

```python
model = YOLO('/kaggle/working/vehicle_classification_model/vehicle_classification_yolov8n/weights/best.pt')
```

```python
results = model.val(
    data='/kaggle/input/vehicle-detection/data/data.yaml',  # Path to your YAML file
    split='test',  # Specify that you want to evaluate on the test set
    imgsz=640,  # Image size
    batch=16,  # Batch size
    project='/kaggle/working/vehicle_classification_model',
    name='test_evaluation'
)
```
